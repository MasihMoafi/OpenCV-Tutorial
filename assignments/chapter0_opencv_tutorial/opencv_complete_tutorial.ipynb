## üì¶ ŸÜÿµÿ® Ÿà Import ⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá‚ÄåŸáÿß

ÿßÿ®ÿ™ÿØÿß ⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá‚ÄåŸáÿß€å ŸÖŸàÿ±ÿØ ŸÜ€åÿßÿ≤ ÿ±ÿß import ŸÖ€å‚Äå⁄©ŸÜ€åŸÖ:

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image, display

# ÿ™ŸÜÿ∏€åŸÖÿßÿ™ matplotlib ÿ®ÿ±ÿß€å ŸÜŸÖÿß€åÿ¥ ÿ®Ÿáÿ™ÿ±
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)

print("‚úì ⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá‚ÄåŸáÿß ÿ®ÿß ŸÖŸàŸÅŸÇ€åÿ™ import ÿ¥ÿØŸÜÿØ")
print(f"OpenCV Version: {cv2.__version__}")
print(f"NumPy Version: {np.__version__}")
````

## üé® ŸÖÿ´ÿßŸÑ 1: ÿß€åÿ¨ÿßÿØ ÿ™ÿµŸà€åÿ± ÿ≥€åÿßŸá

ÿ≥ÿßÿØŸá‚Äåÿ™ÿ±€åŸÜ ⁄©ÿßÿ±: ÿß€åÿ¨ÿßÿØ €å⁄© ÿ™ÿµŸà€åÿ± ÿ≥€åÿßŸá

````python
# ÿß€åÿ¨ÿßÿØ ÿ™ÿµŸà€åÿ± ÿ≥€åÿßŸá 300x300 Ÿæ€å⁄©ÿ≥ŸÑ
black_image = np.zeros((300, 300, 3), dtype=np.uint8)

print(f"ÿ¥⁄©ŸÑ ÿ™ÿµŸà€åÿ±: {black_image.shape}")
print(f"ŸÜŸàÿπ ÿØÿßÿØŸá: {black_image.dtype}")

# ŸÜŸÖÿß€åÿ¥ ÿ™ÿµŸà€åÿ±
plt.imshow(black_image)
plt.title('ÿ™ÿµŸà€åÿ± ÿ≥€åÿßŸá / Black Image')
plt.axis('off')
plt.show()

print("‚úì ÿ™ÿµŸà€åÿ± ÿ≥€åÿßŸá ÿß€åÿ¨ÿßÿØ ÿ¥ÿØ")
````

## üåà ŸÖÿ´ÿßŸÑ 2: ÿß€åÿ¨ÿßÿØ ÿ™ÿµÿßŸà€åÿ± ÿ±ŸÜ⁄Ø€å

OpenCV ÿßÿ≤ ŸÅÿ±ŸÖÿ™ BGR ÿßÿ≥ÿ™ŸÅÿßÿØŸá ŸÖ€å‚Äå⁄©ŸÜÿØ (ŸÜŸá RGB!)

````python
# ÿß€åÿ¨ÿßÿØ ÿ™ÿµÿßŸà€åÿ± ÿ±ŸÜ⁄Ø€å
blue_image = np.zeros((200, 200, 3), dtype=np.uint8)
blue_image[:, :] = (255, 0, 0)  # BGR: ÿ¢ÿ®€å

green_image = np.zeros((200, 200, 3), dtype=np.uint8)
green_image[:, :] = (0, 255, 0)  # BGR: ÿ≥ÿ®ÿ≤

red_image = np.zeros((200, 200, 3), dtype=np.uint8)
red_image[:, :] = (0, 0, 255)  # BGR: ŸÇÿ±ŸÖÿ≤

# ŸÜŸÖÿß€åÿ¥ ÿ™ÿµÿßŸà€åÿ±
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# ÿ™ÿ®ÿØ€åŸÑ BGR ÿ®Ÿá RGB ÿ®ÿ±ÿß€å ŸÜŸÖÿß€åÿ¥ ÿµÿ≠€åÿ≠
axes[0].imshow(cv2.cvtColor(blue_image, cv2.COLOR_BGR2RGB))
axes[0].set_title('ÿ¢ÿ®€å / Blue')
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(green_image, cv2.COLOR_BGR2RGB))
axes[1].set_title('ÿ≥ÿ®ÿ≤ / Green')
axes[1].axis('off')

axes[2].imshow(cv2.cvtColor(red_image, cv2.COLOR_BGR2RGB))
axes[2].set_title('ŸÇÿ±ŸÖÿ≤ / Red')
axes[2].axis('off')

plt.tight_layout()
plt.show()

print("‚úì ÿ™ÿµÿßŸà€åÿ± ÿ±ŸÜ⁄Ø€å ÿß€åÿ¨ÿßÿØ ÿ¥ÿØŸÜÿØ")
print("‚ö†Ô∏è ŸÜ⁄©ÿ™Ÿá: OpenCV ÿßÿ≤ BGR ÿßÿ≥ÿ™ŸÅÿßÿØŸá ŸÖ€å‚Äå⁄©ŸÜÿØÿå ŸÜŸá RGB!")
````

## üìê ŸÖÿ´ÿßŸÑ 3: ÿ±ÿ≥ŸÖ ÿßÿ¥⁄©ÿßŸÑ ŸáŸÜÿØÿ≥€å

OpenCV ÿ™Ÿàÿßÿ®ÿπ ŸÇÿØÿ±ÿ™ŸÖŸÜÿØ€å ÿ®ÿ±ÿß€å ÿ±ÿ≥ŸÖ ÿßÿ¥⁄©ÿßŸÑ ÿØÿßÿ±ÿØ

````python
# ÿß€åÿ¨ÿßÿØ ÿ™ÿµŸà€åÿ± ÿ≥ŸÅ€åÿØ
canvas = np.ones((400, 400, 3), dtype=np.uint8) * 255

# ÿ±ÿ≥ŸÖ ŸÖÿ≥ÿ™ÿ∑€åŸÑ ÿ≥ÿ®ÿ≤
cv2.rectangle(canvas, (50, 50), (150, 150), (0, 255, 0), 3)

# ÿ±ÿ≥ŸÖ ÿØÿß€åÿ±Ÿá ÿ¢ÿ®€å Ÿæÿ± ÿ¥ÿØŸá
cv2.circle(canvas, (300, 100), 50, (255, 0, 0), -1)

# ÿ±ÿ≥ŸÖ ÿÆÿ∑ ŸÇÿ±ŸÖÿ≤
cv2.line(canvas, (50, 250), (350, 250), (0, 0, 255), 2)

# ŸÜŸàÿ¥ÿ™ŸÜ ŸÖÿ™ŸÜ
cv2.putText(canvas, 'OpenCV Tutorial', (80, 350), 
            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

# ŸÜŸÖÿß€åÿ¥
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.title('ÿ±ÿ≥ŸÖ ÿßÿ¥⁄©ÿßŸÑ / Drawing Shapes')
plt.axis('off')
plt.show()

print("‚úì ÿßÿ¥⁄©ÿßŸÑ ÿ±ÿ≥ŸÖ ÿ¥ÿØŸÜÿØ:")
print("  - ŸÖÿ≥ÿ™ÿ∑€åŸÑ ÿ≥ÿ®ÿ≤")
print("  - ÿØÿß€åÿ±Ÿá ÿ¢ÿ®€å")
print("  - ÿÆÿ∑ ŸÇÿ±ŸÖÿ≤")
print("  - ŸÖÿ™ŸÜ")
````

## üíæ ŸÖÿ´ÿßŸÑ 4: ÿ∞ÿÆ€åÿ±Ÿá Ÿà ÿÆŸàÿßŸÜÿØŸÜ ÿ™ÿµŸà€åÿ±

ÿ∞ÿÆ€åÿ±Ÿá ÿ™ÿµŸà€åÿ± ÿØÿ± ŸÅÿß€åŸÑ Ÿà ÿÆŸàÿßŸÜÿØŸÜ ŸÖÿ¨ÿØÿØ ÿ¢ŸÜ

````python
# ÿ∞ÿÆ€åÿ±Ÿá ÿ™ÿµŸà€åÿ±
output_path = 'results/my_first_opencv_image.png'
cv2.imwrite(output_path, canvas)
print(f"‚úì ÿ™ÿµŸà€åÿ± ÿ∞ÿÆ€åÿ±Ÿá ÿ¥ÿØ: {output_path}")

# ÿÆŸàÿßŸÜÿØŸÜ ÿ™ÿµŸà€åÿ±
loaded_image = cv2.imread(output_path)

if loaded_image is not None:
    print(f"‚úì ÿ™ÿµŸà€åÿ± ÿÆŸàÿßŸÜÿØŸá ÿ¥ÿØ")
    print(f"  - ÿßÿ®ÿπÿßÿØ: {loaded_image.shape}")
    print(f"  - ŸÜŸàÿπ ÿØÿßÿØŸá: {loaded_image.dtype}")
    print(f"  - ÿ≠ÿ¨ŸÖ: {loaded_image.nbytes} bytes")
    
    # ŸÜŸÖÿß€åÿ¥
    plt.imshow(cv2.cvtColor(loaded_image, cv2.COLOR_BGR2RGB))
    plt.title('ÿ™ÿµŸà€åÿ± ÿÆŸàÿßŸÜÿØŸá ÿ¥ÿØŸá / Loaded Image')
    plt.axis('off')
    plt.show()
else:
    print("‚úó ÿÆÿ∑ÿß ÿØÿ± ÿÆŸàÿßŸÜÿØŸÜ ÿ™ÿµŸà€åÿ±")
````

## üé® ŸÖÿ´ÿßŸÑ 5: ÿØÿ≥ÿ™⁄©ÿßÿ±€å Ÿæ€å⁄©ÿ≥ŸÑ‚ÄåŸáÿß

ÿ™ÿ∫€å€åÿ± ŸÖÿ≥ÿ™ŸÇ€åŸÖ ŸÖŸÇÿßÿØ€åÿ± Ÿæ€å⁄©ÿ≥ŸÑ‚ÄåŸáÿß

````python
# ÿß€åÿ¨ÿßÿØ ÿ™ÿµŸà€åÿ± ⁄Øÿ±ÿßÿØ€åÿßŸÜÿ™
gradient = np.zeros((256, 256, 3), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        gradient[i, j] = (i, j, 128)  # BGR

# ŸÜŸÖÿß€åÿ¥
plt.imshow(cv2.cvtColor(gradient, cv2.COLOR_BGR2RGB))
plt.title('⁄Øÿ±ÿßÿØ€åÿßŸÜÿ™ ÿ±ŸÜ⁄Ø€å / Color Gradient')
plt.axis('off')
plt.show()

print("‚úì ⁄Øÿ±ÿßÿØ€åÿßŸÜÿ™ ÿ±ŸÜ⁄Ø€å ÿß€åÿ¨ÿßÿØ ÿ¥ÿØ")
````

## üîÑ ŸÖÿ´ÿßŸÑ 6: ÿ™ÿ®ÿØ€åŸÑÿßÿ™ ÿ±ŸÜ⁄Ø€å

ÿ™ÿ®ÿØ€åŸÑ ÿ®€åŸÜ ŸÅÿ∂ÿßŸáÿß€å ÿ±ŸÜ⁄Ø€å ŸÖÿÆÿ™ŸÑŸÅ

````python
# ÿßÿ≥ÿ™ŸÅÿßÿØŸá ÿßÿ≤ ÿ™ÿµŸà€åÿ± ŸÇÿ®ŸÑ€å
original = canvas.copy()

# ÿ™ÿ®ÿØ€åŸÑ ÿ®Ÿá grayscale
gray = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)

# ÿ™ÿ®ÿØ€åŸÑ ÿ®Ÿá HSV
hsv = cv2.cvtColor(original, cv2.COLOR_BGR2HSV)

# ŸÜŸÖÿß€åÿ¥
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
axes[0].set_title('ÿßÿµŸÑ€å / Original (RGB)')
axes[0].axis('off')

axes[1].imshow(gray, cmap='gray')
axes[1].set_title('ÿÆÿß⁄©ÿ≥ÿ™ÿ±€å / Grayscale')
axes[1].axis('off')

axes[2].imshow(cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB))
axes[2].set_title('HSV')
axes[2].axis('off')

plt.tight_layout()
plt.show()

print("‚úì ÿ™ÿ®ÿØ€åŸÑÿßÿ™ ÿ±ŸÜ⁄Ø€å ÿßŸÜÿ¨ÿßŸÖ ÿ¥ÿØ")
````

## üìä ŸÖÿ´ÿßŸÑ 7: ÿßÿ∑ŸÑÿßÿπÿßÿ™ ÿ™ÿµŸà€åÿ±

ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßÿ∑ŸÑÿßÿπÿßÿ™ ŸÖŸÅ€åÿØ ÿßÿ≤ ÿ™ÿµŸà€åÿ±

````python
# ÿßÿ∑ŸÑÿßÿπÿßÿ™ ÿ™ÿµŸà€åÿ±
print("üìä ÿßÿ∑ŸÑÿßÿπÿßÿ™ ÿ™ÿµŸà€åÿ±:")
print(f"  - ÿßÿ®ÿπÿßÿØ: {original.shape}")
print(f"  - ÿßÿ±ÿ™ŸÅÿßÿπ: {original.shape[0]} Ÿæ€å⁄©ÿ≥ŸÑ")
print(f"  - ÿπÿ±ÿ∂: {original.shape[1]} Ÿæ€å⁄©ÿ≥ŸÑ")
print(f"  - ⁄©ÿßŸÜÿßŸÑ‚ÄåŸáÿß: {original.shape[2]}")
print(f"  - ŸÜŸàÿπ ÿØÿßÿØŸá: {original.dtype}")
print(f"  - ÿ≠ÿ¨ŸÖ: {original.nbytes} bytes ({original.nbytes/1024:.2f} KB)")
print(f"  - ÿ™ÿπÿØÿßÿØ Ÿæ€å⁄©ÿ≥ŸÑ: {original.size}")

# ÿ¢ŸÖÿßÿ± Ÿæ€å⁄©ÿ≥ŸÑ‚ÄåŸáÿß
print("\nüìà ÿ¢ŸÖÿßÿ± Ÿæ€å⁄©ÿ≥ŸÑ‚ÄåŸáÿß:")
print(f"  - ŸÖ€åŸÜ€åŸÖŸÖ: {original.min()}")
print(f"  - ŸÖÿß⁄©ÿ≤€åŸÖŸÖ: {original.max()}")
print(f"  - ŸÖ€åÿßŸÜ⁄Ø€åŸÜ: {original.mean():.2f}")
print(f"  - ÿßŸÜÿ≠ÿ±ÿßŸÅ ŸÖÿπ€åÿßÿ±: {original.std():.2f}")
````

## ‚úÇÔ∏è ŸÖÿ´ÿßŸÑ 8: ÿ®ÿ±ÿ¥ Ÿà ÿ™ÿ∫€å€åÿ± ÿßŸÜÿØÿßÿ≤Ÿá

ÿπŸÖŸÑ€åÿßÿ™ ÿ®ÿ±ÿ¥ (crop) Ÿà ÿ™ÿ∫€å€åÿ± ÿßŸÜÿØÿßÿ≤Ÿá (resize)

````python
# ÿ®ÿ±ÿ¥ ÿ™ÿµŸà€åÿ±
cropped = original[50:200, 50:200]

# ÿ™ÿ∫€å€åÿ± ÿßŸÜÿØÿßÿ≤Ÿá
resized_small = cv2.resize(original, (200, 200))
resized_large = cv2.resize(original, (600, 600))

# ŸÜŸÖÿß€åÿ¥
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

axes[0, 0].imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
axes[0, 0].set_title(f'ÿßÿµŸÑ€å / Original ({original.shape[1]}x{original.shape[0]})')
axes[0, 0].axis('off')

axes[0, 1].imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
axes[0, 1].set_title(f'ÿ®ÿ±ÿ¥ ÿ¥ÿØŸá / Cropped ({cropped.shape[1]}x{cropped.shape[0]})')
axes[0, 1].axis('off')

axes[1, 0].imshow(cv2.cvtColor(resized_small, cv2.COLOR_BGR2RGB))
axes[1, 0].set_title(f'⁄©Ÿà⁄Ü⁄© ÿ¥ÿØŸá / Resized Small (200x200)')
axes[1, 0].axis('off')

axes[1, 1].imshow(cv2.cvtColor(resized_large, cv2.COLOR_BGR2RGB))
axes[1, 1].set_title(f'ÿ®ÿ≤ÿ±⁄Ø ÿ¥ÿØŸá / Resized Large (600x600)')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()

print("‚úì ÿπŸÖŸÑ€åÿßÿ™ ÿ®ÿ±ÿ¥ Ÿà ÿ™ÿ∫€å€åÿ± ÿßŸÜÿØÿßÿ≤Ÿá ÿßŸÜÿ¨ÿßŸÖ ÿ¥ÿØ")
````

## üé≠ ŸÖÿ´ÿßŸÑ 9: ŸÖÿßÿ≥⁄© Ÿà ÿπŸÖŸÑ€åÿßÿ™ bitwise

ÿßÿ≥ÿ™ŸÅÿßÿØŸá ÿßÿ≤ ŸÖÿßÿ≥⁄© ÿ®ÿ±ÿß€å ÿßŸÜÿ™ÿÆÿßÿ® ÿ®ÿÆÿ¥‚ÄåŸáÿß€å ÿÆÿßÿµ ÿ™ÿµŸà€åÿ±

````python
# ÿß€åÿ¨ÿßÿØ ŸÖÿßÿ≥⁄© ÿØÿß€åÿ±Ÿá‚Äåÿß€å
mask = np.zeros((400, 400), dtype=np.uint8)
cv2.circle(mask, (200, 200), 150, 255, -1)

# ÿßÿπŸÖÿßŸÑ ŸÖÿßÿ≥⁄©
masked_image = cv2.bitwise_and(original, original, mask=mask)

# ŸÜŸÖÿß€åÿ¥
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
axes[0].set_title('ÿßÿµŸÑ€å / Original')
axes[0].axis('off')

axes[1].imshow(mask, cmap='gray')
axes[1].set_title('ŸÖÿßÿ≥⁄© / Mask')
axes[1].axis('off')

axes[2].imshow(cv2.cvtColor(masked_image, cv2.COLOR_BGR2RGB))
axes[2].set_title('ŸÜÿ™€åÿ¨Ÿá / Result')
axes[2].axis('off')

plt.tight_layout()
plt.show()

print("‚úì ŸÖÿßÿ≥⁄© ÿßÿπŸÖÿßŸÑ ÿ¥ÿØ")
````
