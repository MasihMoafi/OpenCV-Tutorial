## 📦 نصب و Import کتابخانه‌ها

ابتدا کتابخانه‌های مورد نیاز را import می‌کنیم:

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image, display

# تنظیمات matplotlib برای نمایش بهتر
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)

print("✓ کتابخانه‌ها با موفقیت import شدند")
print(f"OpenCV Version: {cv2.__version__}")
print(f"NumPy Version: {np.__version__}")
````

## 🎨 مثال 1: ایجاد تصویر سیاه

ساده‌ترین کار: ایجاد یک تصویر سیاه

````python
# ایجاد تصویر سیاه 300x300 پیکسل
black_image = np.zeros((300, 300, 3), dtype=np.uint8)

print(f"شکل تصویر: {black_image.shape}")
print(f"نوع داده: {black_image.dtype}")

# نمایش تصویر
plt.imshow(black_image)
plt.title('تصویر سیاه / Black Image')
plt.axis('off')
plt.show()

print("✓ تصویر سیاه ایجاد شد")
````

## 🌈 مثال 2: ایجاد تصاویر رنگی

OpenCV از فرمت BGR استفاده می‌کند (نه RGB!)

````python
# ایجاد تصاویر رنگی
blue_image = np.zeros((200, 200, 3), dtype=np.uint8)
blue_image[:, :] = (255, 0, 0)  # BGR: آبی

green_image = np.zeros((200, 200, 3), dtype=np.uint8)
green_image[:, :] = (0, 255, 0)  # BGR: سبز

red_image = np.zeros((200, 200, 3), dtype=np.uint8)
red_image[:, :] = (0, 0, 255)  # BGR: قرمز

# نمایش تصاویر
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# تبدیل BGR به RGB برای نمایش صحیح
axes[0].imshow(cv2.cvtColor(blue_image, cv2.COLOR_BGR2RGB))
axes[0].set_title('آبی / Blue')
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(green_image, cv2.COLOR_BGR2RGB))
axes[1].set_title('سبز / Green')
axes[1].axis('off')

axes[2].imshow(cv2.cvtColor(red_image, cv2.COLOR_BGR2RGB))
axes[2].set_title('قرمز / Red')
axes[2].axis('off')

plt.tight_layout()
plt.show()

print("✓ تصاویر رنگی ایجاد شدند")
print("⚠️ نکته: OpenCV از BGR استفاده می‌کند، نه RGB!")
````

## 📐 مثال 3: رسم اشکال هندسی

OpenCV توابع قدرتمندی برای رسم اشکال دارد

````python
# ایجاد تصویر سفید
canvas = np.ones((400, 400, 3), dtype=np.uint8) * 255

# رسم مستطیل سبز
cv2.rectangle(canvas, (50, 50), (150, 150), (0, 255, 0), 3)

# رسم دایره آبی پر شده
cv2.circle(canvas, (300, 100), 50, (255, 0, 0), -1)

# رسم خط قرمز
cv2.line(canvas, (50, 250), (350, 250), (0, 0, 255), 2)

# نوشتن متن
cv2.putText(canvas, 'OpenCV Tutorial', (80, 350), 
            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

# نمایش
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.title('رسم اشکال / Drawing Shapes')
plt.axis('off')
plt.show()

print("✓ اشکال رسم شدند:")
print("  - مستطیل سبز")
print("  - دایره آبی")
print("  - خط قرمز")
print("  - متن")
````

## 💾 مثال 4: ذخیره و خواندن تصویر

ذخیره تصویر در فایل و خواندن مجدد آن

````python
# ذخیره تصویر
output_path = 'results/my_first_opencv_image.png'
cv2.imwrite(output_path, canvas)
print(f"✓ تصویر ذخیره شد: {output_path}")

# خواندن تصویر
loaded_image = cv2.imread(output_path)

if loaded_image is not None:
    print(f"✓ تصویر خوانده شد")
    print(f"  - ابعاد: {loaded_image.shape}")
    print(f"  - نوع داده: {loaded_image.dtype}")
    print(f"  - حجم: {loaded_image.nbytes} bytes")
    
    # نمایش
    plt.imshow(cv2.cvtColor(loaded_image, cv2.COLOR_BGR2RGB))
    plt.title('تصویر خوانده شده / Loaded Image')
    plt.axis('off')
    plt.show()
else:
    print("✗ خطا در خواندن تصویر")
````

## 🎨 مثال 5: دستکاری پیکسل‌ها

تغییر مستقیم مقادیر پیکسل‌ها

````python
# ایجاد تصویر گرادیانت
gradient = np.zeros((256, 256, 3), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        gradient[i, j] = (i, j, 128)  # BGR

# نمایش
plt.imshow(cv2.cvtColor(gradient, cv2.COLOR_BGR2RGB))
plt.title('گرادیانت رنگی / Color Gradient')
plt.axis('off')
plt.show()

print("✓ گرادیانت رنگی ایجاد شد")
````

## 🔄 مثال 6: تبدیلات رنگی

تبدیل بین فضاهای رنگی مختلف

````python
# استفاده از تصویر قبلی
original = canvas.copy()

# تبدیل به grayscale
gray = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)

# تبدیل به HSV
hsv = cv2.cvtColor(original, cv2.COLOR_BGR2HSV)

# نمایش
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
axes[0].set_title('اصلی / Original (RGB)')
axes[0].axis('off')

axes[1].imshow(gray, cmap='gray')
axes[1].set_title('خاکستری / Grayscale')
axes[1].axis('off')

axes[2].imshow(cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB))
axes[2].set_title('HSV')
axes[2].axis('off')

plt.tight_layout()
plt.show()

print("✓ تبدیلات رنگی انجام شد")
````

## 📊 مثال 7: اطلاعات تصویر

استخراج اطلاعات مفید از تصویر

````python
# اطلاعات تصویر
print("📊 اطلاعات تصویر:")
print(f"  - ابعاد: {original.shape}")
print(f"  - ارتفاع: {original.shape[0]} پیکسل")
print(f"  - عرض: {original.shape[1]} پیکسل")
print(f"  - کانال‌ها: {original.shape[2]}")
print(f"  - نوع داده: {original.dtype}")
print(f"  - حجم: {original.nbytes} bytes ({original.nbytes/1024:.2f} KB)")
print(f"  - تعداد پیکسل: {original.size}")

# آمار پیکسل‌ها
print("\n📈 آمار پیکسل‌ها:")
print(f"  - مینیمم: {original.min()}")
print(f"  - ماکزیمم: {original.max()}")
print(f"  - میانگین: {original.mean():.2f}")
print(f"  - انحراف معیار: {original.std():.2f}")
````

## ✂️ مثال 8: برش و تغییر اندازه

عملیات برش (crop) و تغییر اندازه (resize)

````python
# برش تصویر
cropped = original[50:200, 50:200]

# تغییر اندازه
resized_small = cv2.resize(original, (200, 200))
resized_large = cv2.resize(original, (600, 600))

# نمایش
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

axes[0, 0].imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
axes[0, 0].set_title(f'اصلی / Original ({original.shape[1]}x{original.shape[0]})')
axes[0, 0].axis('off')

axes[0, 1].imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
axes[0, 1].set_title(f'برش شده / Cropped ({cropped.shape[1]}x{cropped.shape[0]})')
axes[0, 1].axis('off')

axes[1, 0].imshow(cv2.cvtColor(resized_small, cv2.COLOR_BGR2RGB))
axes[1, 0].set_title(f'کوچک شده / Resized Small (200x200)')
axes[1, 0].axis('off')

axes[1, 1].imshow(cv2.cvtColor(resized_large, cv2.COLOR_BGR2RGB))
axes[1, 1].set_title(f'بزرگ شده / Resized Large (600x600)')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()

print("✓ عملیات برش و تغییر اندازه انجام شد")
````

## 🎭 مثال 9: ماسک و عملیات bitwise

استفاده از ماسک برای انتخاب بخش‌های خاص تصویر

````python
# ایجاد ماسک دایره‌ای
mask = np.zeros((400, 400), dtype=np.uint8)
cv2.circle(mask, (200, 200), 150, 255, -1)

# اعمال ماسک
masked_image = cv2.bitwise_and(original, original, mask=mask)

# نمایش
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
axes[0].set_title('اصلی / Original')
axes[0].axis('off')

axes[1].imshow(mask, cmap='gray')
axes[1].set_title('ماسک / Mask')
axes[1].axis('off')

axes[2].imshow(cv2.cvtColor(masked_image, cv2.COLOR_BGR2RGB))
axes[2].set_title('نتیجه / Result')
axes[2].axis('off')

plt.tight_layout()
plt.show()

print("✓ ماسک اعمال شد")
````
